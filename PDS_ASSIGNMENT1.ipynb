```python
"""
1.	Create a python program that asks the user how far they want to travel. If they want to travel less than three miles tell them to ride Bicycle.
if they want to travel more than three miles, but less than three hundred miles, tell them to ride Motor-Cycle.
if they want to travel three hundred miles or more tell them to drive Super-Car
"""

d = int(input("How far are you planning to travel (in miles): "))

if d <= 3:
    print("It is best to travel by Bicycle")

elif d <= 300 and d > 3:
    print("It is best to travel by Motor-Cycle")

else:
    print("It is best to travel by Super-car")
```

```python
  """
  2.An App for Mobile hosting provider charges- $0.51 per hour how much does it cost to operator per day, per week, per month?
  how many days can i operate one server with $918?

  """

mhcph = 0.51
mhcpd = mhcph * 24
mhcpw = mhcpd * 7
mhcpm = mhcpw * 4

print(f"Mobile hosting charges per hour = {mhcph}$")
print(f"Mobile hosting charges per day = {mhcpd}$")
print(f"Mobile hosting charges per week = {mhcpw}$")
print(f"Mobile hosting charges per month = {mhcpm}$")

cash = int(input("Input available amount in $: "))
time = cash // mhcpd

print(f"Number of days you can operate with given amount: {time}")
```

```python
"""
4.	A traffic system records the average speed of vehicles every hour for 12 hours.
Write a program that: Reads 12 speed values (floats). Calculates the average speed. Displays whether traffic flow was “Slow” (avg < 40), “Normal” (40–80), or “Fast” (avg > 80).
Concepts: list input, loop, conditional check, average calculation.

"""

speeds = []
print("enter the speed values for the past 12 hours:")

for i in range(12):
    speed = float(input(f"hour {i+1} speed: "))
    speeds.append(speed)

average_speed = sum(speeds) / 12

print(f"\naverage speed = {average_speed}")

if average_speed < 40:
    print("traffic flow: slow")
elif 40 <= average_speed <= 80:
    print("traffic flow: normal")
else:
    print("traffic flow: fast")

```

```python
"""
8.	Output dictionary which contains only the odd numbers that are present in the input list as keys and their cubes as values using List Comprehension

"""

nums = list(map(int, input("Enter numbers separated by spaces: ").split()))

odd_cube_dict = {n: n**3 for n in nums if n % 2 != 0}

print("Dictionary with odd numbers and their cubes:")
print(odd_cube_dict)
```
